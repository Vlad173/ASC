**4. Промежутки знакопостоянства**

Из третьего пункта нам известны точки, в которых функция равна нулю. Можно использовать полученные знания, чтобы определить промежутки знакопостоянства аналитически, подставляя значения x в функцию. _Решим данную задачу графически._ На промежутке (-1; 1) находится график, похожий на параболу, ветви которой стремятся к асимптотам при x = -1 и x = 1, можно сказать, что $\textbf{f(x) > 0}$ **при** $\textbf{x} \in \textbf{(-1; 1)}$ Остальная часть функция расположена ниже оси абсцисс, следовательно $\textbf{f(x) < 0}$ **при** $\textbf{x} \in \textbf{(-2; -1)} \cup \textbf{(1; 2)}$

```python
intervals_of_constancy = plot(0, xmin=-0.1, xmax=0.1, ymin=-0.5, ymax=0.5)

intervals_of_constancy += point( (-0.07, 0), color="black", size=30, zorder=10 ) 
intervals_of_constancy += point( (-0.03, 0), color="gray", size=30, zorder=10 ) 
intervals_of_constancy += point( (0.01, 0), color="gray", size=30, zorder=10 ) 
intervals_of_constancy += point( (0.05, 0), color="black", size=30, zorder=10 ) 

x_margin = 0.08 # смещение по x текста над осью
intervals_of_constancy  += text("$-2$", (-0.07, x_margin), color="black", fontsize=14) 
intervals_of_constancy  += text("$-1$", (-0.03, x_margin), color="black", fontsize=14)  
intervals_of_constancy  += text("$1$", (0.01, x_margin), color="black", fontsize=14) 
intervals_of_constancy  += text("$2$", (0.05, x_margin), color="black", fontsize=14) 

y_margin = 0.1
intervals_of_constancy  += text("-", (-0.05, y_margin), color="red", fontsize=25) 
intervals_of_constancy  += text("+", (-0.01, y_margin), color="green", fontsize=25) 
intervals_of_constancy  += text("-", (0.03, y_margin), color="red", fontsize=25) 

intervals_of_constancy.show(axes=False)
```

**5. Промежутки возрастания и убывания**

Определим промежутки убывания и возрастания по графику. Заметим, что $\textbf{f(x) убывает}$ **при** $\textbf{x} \in \textbf{(-2; -1)} \cup \textbf{(-1; 0)}$ **и** $\textbf{возрастает}$ **при** $\textbf{x} \in \textbf{(0; 1)} \cup \textbf{(1; 2)}$

```python
intervals_of_constancy = plot(0, xmin=-0.1, xmax=0.1, ymin=-0.5, ymax=0.5)

intervals_of_constancy += point( (-0.07, 0), color="black", size=30, zorder=10 ) 
intervals_of_constancy += point( (-0.04, 0), color="gray", size=30, zorder=10 ) 
intervals_of_constancy += point( (-0.01, 0), color="gray", size=30, zorder=10 )
intervals_of_constancy += point( (0.02, 0), color="gray", size=30, zorder=10 ) 
intervals_of_constancy += point( (0.05, 0), color="black", size=30, zorder=10 ) 

x_margin = 0.08 # смещение по x текста над осью
y_margin = 0.1
z_margin = 0.05

intervals_of_constancy  += text("$-2$", (-0.07, x_margin), color="black", fontsize=14) 
intervals_of_constancy  += text("$-1$", (-0.04, x_margin), color="black", fontsize=14)  
intervals_of_constancy  += text("$1$", (0.02, x_margin), color="black", fontsize=14) 
intervals_of_constancy  += text("$2$", (0.05, x_margin), color="black", fontsize=14) 
intervals_of_constancy  += text("$0$", (-0.01, x_margin), color="black", fontsize=14) 

intervals_of_constancy  += arrow((-0.03, y_margin), (-0.015, z_margin), color = 'red') 
intervals_of_constancy  += arrow((0, z_margin), (0.015, y_margin), color = 'red') 
intervals_of_constancy  += arrow((-0.06, y_margin), (-0.045, z_margin), color = 'red') 
intervals_of_constancy  += arrow((0.03, z_margin), (0.045, y_margin), color = 'red') 

intervals_of_constancy.show(axes=False)
```

Когда функция убывает, производная принимает отрицательное значение, когда возрастает - положительное.  
Проверим, подставляя точки из промежутков в производную:

```python
a = y.derivative()
a(-1.5) # производная отрицательна
```

```python
a(-0.5) # производная отрицательна
```

```python
a(0.5) # производная положительна
```

```python
a(1.5) # производная положительна
```

**7. Непрерывность. Наличие точек разрыва и их классификация**

Рассмотрим точки x = -1 и x = 1, найдём левосторонние и правосторонние пределы для этих точек.

```python
show(limit(y, x = -1, dir='-'))
show(limit(y, x = -1, dir = '+'))
show(limit(y, x = 1, dir = '+'))
show(limit(y, x = 1, dir = '-'))
```

Все односторонние пределы бесконечны, значит функция непрырывна на всей числовой прямой кроме точек x = 1 и x = -1, в которых она терпит **разрывы второго рода.**

**8. Асимптоты**

Из пункта семь следует, что прямые x = -1, x = 1 являются вертикальными асимптотами для графика данной функции, так как в этих точках односторонние пределы равны бесконечности.

```python
g = plot(y, ymin = -5, ymax = 5, xmin = -5, xmax = 5, aspect_ratio = 1, exclude = [-1, 1])
g += line([(-1, -5), (-1, 5)], linestyle = 'dotted', color = 'red')
g += line([(1, -5), (1, 5)], linestyle = 'dotted', color = 'red')
g.show()
```
